```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import html5lib
```

```python
lista = []
for i in range(1,101, 50) :
   
    url2 = f'https://www.imdb.com/search/title/?title_type=feature&num_votes=10000,&sort=user_rating,desc&start={i}&ref_=adv_nxt'
    link1 = requests.get(url2)
    sopa = BeautifulSoup(link1.content)
    hack = sopa.select('h3.lister-item-header')
    lista.append([tag.text for tag in hack])
flat_list = [item for sublist in lista for item in sublist]

print('done')
```
